```python
df_redwinequality_new = df_redwinequality[['volatile acidity','sulphates','citric acid','alcohol','residual sugar','fixed acidity','chlorides','free sulfur dioxide','total sulfur dioxide','pH','quality']].copy()
```

```python
df_redwinequality_new = df_redwinequality_new.rename(columns = {'volatile acidity': 'volatile_acidity', 'citric acid': 'citric_acid','residual sugar':'residual_sugar', 'fixed acidity':'fixed_acidity','free sulfur dioxide':'free_sulfur_dioxide','total sulfur dioxide':'total_sulfur_dioxide'}, inplace = False)
df_redwinequality_new.head()
```

```python
df_redwinequality_new['quality_label'] = df_redwinequality_new.quality.apply(lambda q: '1' if q <= 5 else '2' if q < 7 else '3')
df_redwinequality_new.head()
```

```python
ls = round(df_redwinequality_new[df_redwinequality_new['quality_label'] == '1'].describe(),2)
ms = round(df_redwinequality_new[df_redwinequality_new['quality_label'] == '2'].describe(),2)
hs = round(df_redwinequality_new[df_redwinequality_new['quality_label'] == '3'].describe(),2)
pd.concat([ls, ms, hs], axis=0, keys=['Low Quality Wine', 'Medium Quality Wine', 'High Quality Wine']).T
```

```python
from scipy import stats
r,p = stats.pearsonr(df_redwinequality_new.quality, df_redwinequality_new.sulphates)
#print(round(p,25))
print('The p-value for sulphates and quality is',(round(p,25)), 'which is less than 0.05; therefore the relationship between the two variables is statistically significant')
```

```python
r,p = stats.pearsonr(df_redwinequality_new.quality, df_redwinequality_new.volatile_acidity)
#print(p)
print('The p-value for volatile_acidity and quality is',(round(p,61)), 'which is less than 0.05; therefore the relationship between the two variables is statistically significant')
```
