```python
redwinequality = "Resources/winequality-red.csv"
```

```python
df_redwinequality = pd.read_csv(redwinequality, encoding="ISO-8859-1")
```

```python
df_redwinequality.head(100)
```

```python
df_redwinequality.count()
```

```python
df_redwinequality = df_redwinequality.dropna(how='any')
```

```python
df_redwinequality.count()
```

```python
df_redwinequality.corr()
```

```python
df_redwinequality_new = df_redwinequality[['volatile acidity','sulphates','citric acid','alcohol','residual sugar','fixed acidity','chlorides','free sulfur dioxide','pH','quality']].copy()
```

```python
df_redwinequality_new = df_redwinequality_new.rename(columns = {'volatile acidity': 'volatile_acidity', 'citric acid': 'citric_acid','residual sugar':'residual_sugar', 'fixed acidity':'fixed_acidity','free sulfur dioxide':'free_sulfur_dioxide'}, inplace = False)
df_redwinequality_new.head()
```

```python
df_redwinequality_new['quality_label'] = df_redwinequality_new.quality.apply(lambda q: '1' if q <= 5 else '2' if q < 7 else '3')
df_redwinequality_new.head()
```

```python
from scipy import stats
r,p = stats.pearsonr(df_redwinequality_new.quality, df_redwinequality_new.sulphates)
#print(round(p,25))
print('The p-value for sulphates and quality is',(round(p,25)), 'which is less than 0.05; therefore the relationship between the two variables is statistically significant')
```

```python
r,p = stats.pearsonr(df_redwinequality_new.quality, df_redwinequality_new.volatile_acidity)
#print(p)
print('The p-value for volatile_acidity and quality is',(round(p,61)), 'which is less than 0.05; therefore the relationship between the two variables is statistically significant')
```

```python
r,p = stats.pearsonr(df_redwinequality_new.quality, df_redwinequality_new.citric_acid)
#print(p)
print('The p-value for citric_acid and quality is',(round(p,22)), 'which is less than 0.05; therefore the relationship between the two variables is statistically significant')
```

```python
r,p = stats.pearsonr(df_redwinequality_new.quality, df_redwinequality_new.alcohol)
print(p)
print('The p-value for alcohol and quality is',(round(p,93)), 'which is less than 0.05; therefore the relationship between the two variables is statistically significant')
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
 X= df_redwinequality_new[['volatile_acidity', 'sulphates', 'citric_acid', 'alcohol','residual_sugar','fixed_acidity','chlorides','free_sulfur_dioxide','pH']]
```

```python
y=df_redwinequality_new[['quality_label']]
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

# Fitting our model with all of our features in X
model.fit(X, y)

score = model.score(X, y)
print(f"R2 Score: {score}")
print('The R2 Score for the model is',(round(score,6)),', therefore, the model explains > 36% of the variability of the response data around its mean.') 
```

```python
fig = plt.figure(figsize = (18, 4))
title = fig.suptitle("Red Wine Quality", fontsize=14)
fig.subplots_adjust(top=0.85, wspace=0.3)

ax1 = fig.add_subplot(1,4, 1)
ax1.set_title("Red Wine")
ax1.set_xlabel("Quality")
ax1.set_ylabel("Frequency") 
rw_q = df_redwinequality_new.quality.value_counts()
rw_q = (list(rw_q.index), list(rw_q.values))
ax1.set_ylim([0,1000])
ax1.tick_params(axis='both', which='major', labelsize=8.5)
bar1 = ax1.bar(rw_q[0], rw_q[1], color='red', edgecolor='black', linewidth=1)

ax3 = fig.add_subplot(1,4, 3)
ax3.set_title("Red Wine")
ax3.set_xlabel("Quality Label")
ax3.set_ylabel("Frequency") 
rw_q = df_redwinequality_new.quality_label.value_counts()
rw_q = (list(rw_q.index), list(rw_q.values))
ax3.set_ylim([0, 1000])
bar3 = ax3.bar(list(range(len(rw_q[0]))), rw_q[1], color='red', edgecolor='black', linewidth=1, tick_label =rw_q[0])
```
